### POPULATE PROPERTY ADDRESS DATA (NULL/MISSING DATA)
Translating a SQL self-join in Pandas. Not the most efficient way but I did this to help me understand and learn better.

An alternative way of doing this would be to do: 
- GroupBy
- Helper function to return missing values for each group
- Transform

```python
#Checking where we have null values
    #df.isnull().sum()
#Self-joining the data
df_copy = df.merge(df, on = 'ParcelID', suffixes = ('_left', '_right'))
df_copy
```

```python
#Realized I have a space in the column header, so removing them
df.columns = df.columns.str.replace(" ", "")
df.columns
```

```python
#Removes rows where the merge paired a row with itself. (SQL tb.UniqueID <> tb_2.UniqueID.)
df_copy = df_copy[df_copy["UniqueID_left"] != df_copy["UniqueID_right"]]
df_copy
```

```python
#Now we only want to keep pairs where the left-side address is NULL and the right-side address is non-NULL.
# This is beacsue we want to insert the values from the right side dataframe to the missing left side ones.
df_copy = df_copy[df_copy["PropertyAddress_left"].isna() & df_copy["PropertyAddress_right"].notna()]
df_copy
```

```python
#Creating a 1:1 map dict, to accurately mapp the right value to the PropertyAddress
    #Mapping the left uniqueID key to the value of PropertyAddress_right
fill_map = df_copy.set_index("UniqueID_left")["PropertyAddress_right"].to_dict()
fill_map
```

```python
#Updating the PropertyAddress with the map we created above.

#update_rows selects exactly the rows you intend to update.
update_rows = df["UniqueID"].isin(fill_map.keys())
# Filling the missing PropertyAddress cells with the corresponding mapped values.
df.loc[update_rows, "PropertyAddress"] = df.loc[update_rows, "UniqueID"].map(fill_map) #Matches the UniqueID keys with the keys in filled_map and inserts the mapped values.
df
```

```python
# Double checking the addresses in df matches the mapped ones (in fill_map)
df.loc[update_rows, "PropertyAddress"].isin(fill_map.values())
# Double checking so that the ID's in the df matches the mapped id's.
df.loc[update_rows, "UniqueID"].isin(fill_map.keys())

#checking for null values, there are no longer any null values in the PropertyAddress
df.isnull().sum()
```
